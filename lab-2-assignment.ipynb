### 3. Create a line plot using Matplotlib where the x-axis represents the letters of your name (e.g., ['J', 'O', 'H', 'N']) and the y-axis represents the number of times each letter appears in your name.

```python
import matplotlib.pyplot as plt
from collections import Counter

name = "Aayusha"

# Count the occurrences of each letter in the name
letter_counts = Counter(name)

# Extract the letters and their counts
letters = list(letter_counts.keys())
counts = list(letter_counts.values())

# Create the line plot
plt.plot(letters, counts, marker='o')

# Set the title and labels
plt.title("Letter Frequency in My Name")
plt.xlabel("Letters")
plt.ylabel("Frequency")

# Display the plot
plt.show()
```

### 10. Create a scatter plot with a regression line using Seaborn. Use random x values and calculate y values using the formula y = 2x + noise. Title the chart with your name (e.g., "John's Scatter Plot with Regression Line")

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generate 50 random x values
x = np.random.rand(50) * 10  # Random values between 0 and 10

# Calculate y values using the formula y = 2x + noise
noise = np.random.randn(50)  # Random noise
y = 2 * x + noise  # y = 2x + noise

# Create a scatter plot with a regression line
sns.regplot(x=x, y=y, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})

# Add a title to the plot with your name
plt.title("Anshu's Scatter Plot with Regression Line")

# Label the axes
plt.xlabel("X Values")
plt.ylabel("Y Values")

# Display the plot
plt.show()
```
